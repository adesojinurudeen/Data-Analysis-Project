```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')  
import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('branches.csv')
```

```python
sns.catplot(x="Product line", y="gross income", data=df, aspect=4)
# to analyse the product line and gross income we used catplot to know the product line that generate more income.
#and looking at chart below, fashion accessories generates most income 
```

```python
sns.catplot(x='Product line',y='Quantity', Units='Product line per Unit price', data=df, aspect=4, kind= 'point')
#Using catplot visualisation to analyse product line, quantity and unit price per product. which indicate that product 
#with highest unit price is Electronic accessories, follow by home and lifestyle and the lowest unit price is Fashion accessories.
```

```python
sns.catplot(x='Product line',y='Quantity', Units='product line per Quantity', data=df, aspect=4, kind='point')
#Using catplot visualisation to analyse product line, quantity and the quantity per product. which indicate that product 
#with highest quantity is Electronic accessories, follow by home and lifestyle.
```

```python
sns.countplot(y='Product line', hue='Payment', data=df).set_title("Product analysis")
#this graph also output the product line with highest payment method.
#this indicate Electronic accessories to be the highest product purchased
#with cash followed by fashion accessories and home and lifestyle 
```

```python
sns.boxplot(x='Branch', y='gross income', data=df).set_title("Sales Records")
# this boxplot visualization indicates the branch with most gross income.
```

```python
sns.catplot(x='Product line', y='Total', hue='Gender', data=df, aspect=4, kind= 'point')
#Using catplot visualisation to analyse product line, Total and patronage base on gender. 
# The chat below indicate female patronize food and beverages, home and lifestyle much more than men, 
# and male patronize health and beauty, sports and travel much more than female
```

```python
sns.countplot(x='Payment', hue='Branch', data=df).set_title("Sales Records")
# this countplot shows the most payment made by branch. which this graph points at branch A to be leading at Epay. 
```

```python
#Using a catplot() generate visualization for the "product line" on x-axis, quantity on the y-axis, and hue as gender.
sns.catplot(x='Product line', y='Quantity', hue='Gender', data=df, aspect=4) #.set_title("Sales Records")
```

```python
sns.countplot(x="Branch", hue= 'Rating', data=df, order= df['Branch']).set_title("Rating across Branches")
# using countplot to visualize  analysis of branch and Rating. 
#the chat below shows the branch with highest and lowest rating.
```
